```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.svm import SVC, SVR
from sklearn.metrics import classification_report, accuracy_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline
file_path = r'C:\Users\sv697\Downloads\avocado.csv\avocado.csv'
avocado_data = pd.read_csv(file_path)

classification_features = ['AveragePrice', 'Total Volume', '4046', '4225', '4770']
X_class = avocado_data[classification_features]
y_class = avocado_data['type']

X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_class, y_class, test_size=0.2, random_state=42)

scaler_class = StandardScaler()
X_train_class = scaler_class.fit_transform(X_train_class)
X_test_class = scaler_class.transform(X_test_class)

log_reg = LogisticRegression(random_state=42)
log_reg.fit(X_train_class, y_train_class)

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train_class, y_train_class)

svm_classifier = SVC(random_state=42)
svm_classifier.fit(X_train_class, y_train_class)

models = [log_reg, rf_classifier, svm_classifier]
model_names = ['Logistic Regression', 'Random Forest Classifier', 'SVM Classifier']

for i, model in enumerate(models):
    cv_scores = cross_val_score(model, X_train_class, y_train_class, cv=5)
    print(f"\n{model_names[i]} Cross-Validation Scores:", cv_scores)
    print(f"Mean Cross-Validation Score: {np.mean(cv_scores):.4f}")

    y_pred_class = model.predict(X_test_class)

    print(f"\n{model_names[i]} Classification Report:\n", classification_report(y_test_class, y_pred_class))
    print(f"{model_names[i]} Accuracy:", accuracy_score(y_test_class, y_pred_class))

regression_features = ['Total Volume', '4046', '4225', '4770']
X_reg = avocado_data[regression_features]
y_reg = avocado_data['AveragePrice']

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

scaler_reg = StandardScaler()
X_train_reg = scaler_reg.fit_transform(X_train_reg)
X_test_reg = scaler_reg.transform(X_test_reg)

reg_model = LinearRegression()
reg_model.fit(X_train_reg, y_train_reg)

rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor.fit(X_train_reg, y_train_reg)

svm_regressor = SVR()
svm_regressor.fit(X_train_reg, y_train_reg)

reg_models = [reg_model, rf_regressor, svm_regressor]
reg_model_names = ['Linear Regression', 'Random Forest Regressor', 'SVM Regressor']

for i, reg_model in enumerate(reg_models):
    cv_scores_reg = cross_val_score(reg_model, X_train_reg, y_train_reg, cv=5, scoring='neg_mean_squared_error')
    print(f"\n{reg_model_names[i]} Cross-Validation Scores:", -cv_scores_reg)
    print(f"Mean Cross-Validation Score: {np.mean(-cv_scores_reg):.4f}")

    y_pred_reg = reg_model.predict(X_test_reg)

    print(f"\n{reg_model_names[i]} Mean Absolute Error:", mean_absolute_error(y_test_reg, y_pred_reg))
    print(f"{reg_model_names[i]} Mean Squared Error:", mean_squared_error(y_test_reg, y_pred_reg))

    plt.scatter(y_test_reg, y_pred_reg)
    plt.xlabel("Actual Prices")
    plt.ylabel("Predicted Prices")
    plt.title(f"{reg_model_names[i]}: Actual Prices vs. Predicted Prices")
    plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.svm import SVC, SVR
from sklearn.metrics import classification_report, accuracy_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline
file_path = r'C:\Users\sv697\Downloads\avocado.csv\avocado.csv'
avocado_data = pd.read_csv(file_path)
classification_features = ['Total Volume', '4046', '4225', '4770']
X_class = avocado_data[classification_features]
y_class = avocado_data['type']

X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_class, y_class, test_size=0.2, random_state=42)

# Standardize the features
scaler_class = StandardScaler()
X_train_class = scaler_class.fit_transform(X_train_class)
X_test_class = scaler_class.transform(X_test_class)

rf_classifier = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search_class = GridSearchCV(rf_classifier, param_grid, cv=5)
grid_search_class.fit(X_train_class, y_train_class)

best_params_class = grid_search_class.best_params_
print("Best Hyperparameters for Random Forest Classifier:", best_params_class)

best_rf_classifier = grid_search_class.best_estimator_
cv_scores_class = cross_val_score(best_rf_classifier, X_train_class, y_train_class, cv=5)
print("Best Random Forest Classifier Cross-Validation Scores:", cv_scores_class)
print("Mean Cross-Validation Score:", np.mean(cv_scores_class))

y_pred_class = best_rf_classifier.predict(X_test_class)

print("Classification Report:\n", classification_report(y_test_class, y_pred_class))
print("Accuracy:", accuracy_score(y_test_class, y_pred_class))

regression_features = ['Total Volume', '4046', '4225', '4770']
X_reg = avocado_data[regression_features]
y_reg = avocado_data['AveragePrice']

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

scaler_reg = StandardScaler()
X_train_reg = scaler_reg.fit_transform(X_train_reg)
X_test_reg = scaler_reg.transform(X_test_reg)

rf_regressor = RandomForestRegressor(random_state=42)

param_grid_reg = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search_reg = GridSearchCV(rf_regressor, param_grid_reg, cv=5, scoring='neg_mean_squared_error')
grid_search_reg.fit(X_train_reg, y_train_reg)

# Best hyperparameters
best_params_reg = grid_search_reg.best_params_
print("\nBest Hyperparameters for Random Forest Regressor:", best_params_reg)

# Evaluate the best model
best_rf_regressor = grid_search_reg.best_estimator_
cv_scores_reg = cross_val_score(best_rf_regressor, X_train_reg, y_train_reg, cv=5, scoring='neg_mean_squared_error')
print("Best Random Forest Regressor Cross-Validation Scores:", -cv_scores_reg)
print("Mean Cross-Validation Score:", np.mean(-cv_scores_reg))

y_pred_reg = best_rf_regressor.predict(X_test_reg)

print("\nMean Absolute Error:", mean_absolute_error(y_test_reg, y_pred_reg))
print("Mean Squared Error:", mean_squared_error(y_test_reg, y_pred_reg))

plt.scatter(y_test_reg, y_pred_reg)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Random Forest Regressor: Actual Prices vs. Predicted Prices")
plt.show()
```
